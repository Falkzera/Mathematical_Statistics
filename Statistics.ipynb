# Filtragem de dados
Verificando os dtypes dos dados.

```python
data.dtypes
```

### Observe os tipos de dados:
Veja que 'DATA INICIAL' e 'DATA FINAL' estão no formato: 'Object'. Deveriam estar no formato 'Datetime'

```python
# Transformando para Datetime
data['DATA INICIAL'] = pd.to_datetime(data['DATA INICIAL']) 
data['DATA FINAL'] = pd.to_datetime(data['DATA FINAL'])
data.dtypes
```

Observe que existe alguns valores númericos, que estão no formato 'Object'
Valores númericos, em sua maioria, devem representar valores Inteiros ou float.
Nesse caso, iremos converter para Float os valores que estão como objetc

```python
'''
O código abaixo cria uma lista com os nomes das colunas que serão modificadas,
 e em seguida, ele percorre a lista e substitui os valores '-' por NaN e em seguida transforma os valores em float.
'''

colunas = [
    'MARGEM MÉDIA REVENDA', 
    'PREÇO MÉDIO DISTRIBUIÇÃO', 
    'DESVIO PADRÃO DISTRIBUIÇÃO', 
    'PREÇO MÍNIMO DISTRIBUIÇÃO', 
    'PREÇO MÁXIMO DISTRIBUIÇÃO', 
    'COEF DE VARIAÇÃO DISTRIBUIÇÃO'
]

for coluna in colunas:
    data[coluna] = data[coluna].replace('-', np.nan).astype(float)

data.dtypes
```

### Filtragem de dados - parte 1
Antes de iniciarmos com a parte estatística, iremos primeiro filtrar os dados para o foco da análise.
Objetivo é análisar o preço dos produtos da região Nordeste de 2018 até 2021.
Assim, estaremos inseridos no período antes e durante a pandemia do Covid-19.

Observe a syntaxe utilizada, ela será a mesma para qualquer outra filtragem.

```python
data_filtrado = data[data['REGIÃO'] == 'NORDESTE'] # Criando uma nova variavel de dataframe filtrado.
data_filtrado.shape # Observe a redução de linhas. Essa redução é proveniente do filtro apenas para a região Nordeste.
```

```python
data_filtrado.head(3) # Exibindo o dataframe filtrado para a região.
```

### Filtragem de dados - parte 2
Filtrando agora para o período de análise. [2018 - 2021]

```python
# Criando um filtro de 2018 até 2021
data_filtrado = data_filtrado[data_filtrado['DATA INICIAL'] >= '2018-01-01']
data_filtrado.shape
```

### Filtragem de dados - parte 3
Com o Dataframe pronto para ser análisado, iremos agora resetar o indíce para alinhar o númericamente os dados.

```python
# Resetando o index do dataframe.
data_filtrado.reset_index(drop=True, inplace=True) # Drop=True para não criar uma nova coluna com o index antigo. inplace=True para modificar o dataframe original.
data_filtrado.head()
```

### Filtragem de dados - parte 4
Adicionando uma última filtragem, as únicas colunas de dados que irá nós interessar para esse estudo é a coluna de Preço médio de distribuição e Preço médio de revenda.

```python
# Limpando o Dataframe e filtrando apenas os dados de interesse
data_filtrado = data_filtrado[['DATA INICIAL', 'DATA FINAL', 'PRODUTO', 'ESTADO', 'PREÇO MÉDIO REVENDA']]
data_filtrado.shape # Observe a diminuição de linhas e colunas, tornando um Dataframe mais limpo e objetivo.
```

```python
data_filtrado
```

### Filtragem de dados - parte 5
Escolhendo apenas o PRODUTO: Gasolina Comum para análise e criando uma nova variável.

```python
# Filtrando apenas os dados de Gasolina Comum
data_filtrado_gasolina = data_filtrado[data_filtrado['PRODUTO'] == 'GASOLINA COMUM'].copy()
# Definindo apenas 2 casas decimais para PREÇO MÉDIO REVENDA e PREÇO MÉDIO DISTRIBUIÇÃO
data_filtrado_gasolina['PREÇO MÉDIO REVENDA'] = data_filtrado_gasolina['PREÇO MÉDIO REVENDA'].round(2)
data_filtrado.reset_index(drop=True, inplace=True)
data_filtrado_gasolina
```

# Análise Estatística
Com o DataFrame devidamente configurado, agora podemos iniciar a análise.

### Medidas de tendência central  
Os métodos (method) mais comuns são:  

**mean()**: Cálcula a média arítimética.  
**median()**: Encontra o valor mediano.  
**quantile([])**: Cálcula os quantis dos dados, que são pontos de cortes que dividem os dados em intervalos de distribuições.  
**mode()**: Retorna o valor que mais aparece no Dataframe.  

```python
data_filtrado_gasolina['PREÇO MÉDIO REVENDA'].mean() # Calculando a média.
```

```python
data_filtrado_gasolina['PREÇO MÉDIO REVENDA'].median() # Calculando a mediana.
```

```python
data_filtrado_gasolina['PREÇO MÉDIO REVENDA'].quantile([0.0, 0.25, 0.50, 0.75, 1.0]) # Calculando os quartis.
```

```python
data_filtrado_gasolina['PREÇO MÉDIO REVENDA'].mode() # Verificando a moda.
```

### Medidas de dispersão e distribuição

```python
data_filtrado_gasolina['PREÇO MÉDIO REVENDA'].quantile([0,1]) # Verificando o valor mínimo e máximo.
# Observa-se também a amplitude, a distância entre o valor máximo e o valor mínimo.
```

```python
data_filtrado_gasolina['PREÇO MÉDIO REVENDA'].max() # O valor máximo também pode ser obtido através desse método.
```

```python
data_filtrado_gasolina['PREÇO MÉDIO REVENDA'].min() # O valor mínimo também pode ser obtido através desse método.
```

```python
data_filtrado_gasolina['PREÇO MÉDIO REVENDA'].max() - data_filtrado_gasolina['PREÇO MÉDIO REVENDA'].min() # Calculando a amplitude.
```

```python
data_filtrado_gasolina['PREÇO MÉDIO REVENDA'].describe() # Descreve de maneira direta os dados estatísticos.
```

#### Interquatile Range (IQR)
Obtido através da função np.nanpercitile([''])

# COMPARANDOS REGIÕES
Adicionando os mesmo filtros, porém mantendo todas as Regiões e todos os Estados.

```python
data_filter = data = data[['DATA INICIAL', 'DATA FINAL', 'PRODUTO', 'ESTADO', 'REGIÃO', 'PREÇO MÉDIO REVENDA']]
data_filter = data_filter[data_filter['PRODUTO'] == 'GASOLINA COMUM']
data_filter.reset_index(drop=True, inplace=True)
data_filter.shape
```

### Observando Gráficamente os dados 
3.3 Gráfico de linha:

Nesse novo Dataframe a análise é de 2004 até 2021, incluindo todas as regiões unificadas

```python
plt.figure(figsize=(20, 12))
sns.lineplot(data=data_filter, x='DATA INICIAL', y='PREÇO MÉDIO REVENDA', errorbar=None)
plt.title('Preço Médio de Revenda ao Longo do Tempo')
plt.xlabel('Data Inicial')
plt.ylabel('Preço Médio de Revenda')
plt.xticks(rotation=45)
plt.show()
```

### Observando Gráficamente os dados 
3.3 Gráfico de linha:

Nesse novo Dataframe a análise é de 2004 até 2021, incluindo todas as regiões separadas

```python
plt.figure(figsize=(20, 12))
sns.lineplot(data=data_filter, x='DATA INICIAL', y='PREÇO MÉDIO REVENDA', hue='REGIÃO', errorbar=None)
plt.title('Preço Médio de Revenda ao Longo do Tempo')
plt.xlabel('Data Inicial')
plt.ylabel('Preço Médio de Revenda')
plt.xticks(rotation=45)
plt.show()
```

### Observando Gráficamente os dados 
3.4 Gráfico de linha:
#### Comparação: Sudeste x Nordeste
de 2004 até 2021

```python
# Filtrando apenas duas regiões: Nordeste e Sudeste
nordeste_sudeste = data_filter[(data_filter['REGIÃO'] == 'NORDESTE') | (data_filter['REGIÃO'] == 'SUDESTE')]
nordeste_sudeste.reset_index(drop=True, inplace=True)

# Comparar os dados entre regiões
plt.figure(figsize=(20, 12))
sns.lineplot(data=nordeste_sudeste, x='DATA INICIAL', y='PREÇO MÉDIO REVENDA', hue='REGIÃO', errorbar=None)
plt.title('Preço Médio de Revenda ao Longo do Tempo')
plt.xlabel('Data Inicial')
plt.ylabel('Preço Médio de Revenda')
plt.xticks(rotation=45)
plt.show()
```

### Observando Gráficamente os dados 
3.5 Gráfico Boxplot
#### Comparação: Sudeste x Nordeste

```python
plt.figure(figsize=(20, 12))
sns.boxplot(x='REGIÃO', y='PREÇO MÉDIO REVENDA', data=nordeste_sudeste)
plt.title('Comparação dos Preços Médios de Revenda entre Nordeste e Sudeste')
plt.show()
```

### Observando Gráficamente os dados 
3.6 Gráfico violino
#### Comparação: Sudeste x Nordeste

```python
plt.figure(figsize=(20, 12))
sns.violinplot(x='REGIÃO', y='PREÇO MÉDIO REVENDA', data=nordeste_sudeste, inner='quartile')
plt.title('Distribuição dos Preços Médios de Revenda por Região')
plt.show()
```

### Observando Gráficamente os dados 
3.7 Histograma
#### Comparação: Sudeste x Nordeste

```python
plt.figure(figsize=(20, 12))
sns.histplot(data=nordeste_sudeste, x='PREÇO MÉDIO REVENDA', hue='REGIÃO', element='step', stat='density', common_norm=False)
plt.title('Distribuição dos Preços Médios de Revenda')
plt.show()
```
